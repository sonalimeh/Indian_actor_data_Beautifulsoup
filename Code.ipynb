```
from requests import get
url = 'https://www.imdb.com/search/name/?birth_place=India&bio=Indian&has=trivia'
urls=[]
urls.append(url)
for k in range(51,302,50):
  nextp='&start='+str(k)+'&ref_=rlm'
  urls.append(url+nextp)
response = get(url)
print(response.text[:500])
```

```
from bs4 import BeautifulSoup
name=[]
dis=[]
img=[]
for i in urls:
  response = get(i)
  html_soup = BeautifulSoup(response.text, 'html.parser')
  movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-detail')
  print(type(movie_containers))
  print(len(movie_containers))
  for cont in movie_containers:
    name.append(cont.h3.a.text)
    dis.append(cont.find("p").findNext("p").get_text())
    img.append(cont.find('img',src=True))
```

```
fin=[]
!pip install wget
import wget
for link in img:
    link = link["src"].split("src=")[-1]
    local_image_filename = wget.download(link)
    fin.append(local_image_filename)
print(fin)
    
```

```
import pickle
from sklearn.feature_extraction.text import CountVectorizer
import plotly.express as px
import pandas as pd
import re
```

```
cEXT = pickle.load( open( "/content/drive/My Drive/data/models/cEXT.p", "rb"))
cNEU = pickle.load( open( "/content/drive/My Drive/data/models/cNEU.p", "rb"))
cAGR = pickle.load( open( "/content/drive/My Drive/data/models/cAGR.p", "rb"))
cCON = pickle.load( open( "/content/drive/My Drive/data/models/cCON.p", "rb"))
cOPN = pickle.load( open( "/content/drive/My Drive/data/models/cOPN.p", "rb"))
vectorizer_31 = pickle.load( open( "/content/drive/My Drive/data/models/vectorizer_31.p", "rb"))
vectorizer_30 = pickle.load( open( "/content/drive/My Drive/data/models/vectorizer_30.p", "rb"))
```

```
def predict_personality(text):
    scentences = re.split("(?<=[.!?]) +", text)
    text_vector_31 = vectorizer_31.transform(scentences)
    text_vector_30 = vectorizer_30.transform(scentences)
    EXT = cEXT.predict(text_vector_31)
    NEU = cNEU.predict(text_vector_30)
    AGR = cAGR.predict(text_vector_31)
    CON = cCON.predict(text_vector_31)
    OPN = cOPN.predict(text_vector_31)
    return [EXT[0], NEU[0], AGR[0], CON[0], OPN[0]]
```

```
disk=[]
```

```
f=[ 'Extraversion','Neuroticism', 'Agreeableness','Conscientiousness','Openness']
for i in dis:
  str1=''
  predictions = predict_personality(i)
  #print("predicted personality:", predictions)
  for j in range(0, len(predictions)):
    if(predictions[j]==1):
      str1=str1+' '+f[j]
  disk.append(str1)

```
